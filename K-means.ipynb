# **Market Segmentation**


> We have a data of mall customers. We are finding the clusters of customers having similar shoping traits. We are using "Annual Income (k$)" and "Spending Score" (which is derived from the perivous purchases) to found similarities between group of customers. We are using "**K-Means**" for cluster analysis.






# K-Means

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Get Dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, 3:].values
```

### Using Elbow Method to Find Out optimal Number of Clustres

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init = 10, max_iter = 300, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Fitting K -Means To Dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

### Visualising K - Means on Cuatomers Data

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 60, c = 'red', label = 'careful')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 60, c = 'blue', label = 'Standard')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 60, c = 'green', label = 'Target')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 60, c = 'cyan', label = 'careless')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 60, c = 'magenta', label = 'sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 90, c = 'black', label = 'centroids') 
plt.title('Clusters of Customers')
plt.xlabel('Annual Salary (k$)')
plt.ylabel('Spending Score (1 - 100)')
plt.legend()
plt.show()
```
