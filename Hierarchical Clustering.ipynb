# **Market Segmentation**


> We have a data of mall customers. We are finding the clusters of customers having similar shoping traits. We are using "Annual Income (k$)" and "Spending Score" (which is derived from the perivous purchases) to found similarities between group of customers. We are using "**Hierarchical Clustering**" for cluster analysis.


# Hierarchical Clustering

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Get the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, 3:].values
```

### Using Dendrogram to Find Optimal Number of Clusters

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

### Fitting Hierarchical (Agglomerative) Cluster to Dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x)
```

### Visualising Clusters

```python
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 60, c = 'red', label = 'Careful')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 60, c = 'blue', label = 'Standard')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 60, c = 'green', label = 'Target')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 60, c = 'cyan', label = 'Careless')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 60, c = 'magenta', label = 'Sensible')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
